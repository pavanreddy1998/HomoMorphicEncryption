```
import numpy as np
from PIL import Image

icon_circle = Image.open('circle.png').convert('RGBA')
arr_circle = np.array(icon_circle)

# record the original shape
shape_circle = arr_circle.shape

# make a 1-dimensional view of arr
flat_arr_circle = arr_circle.ravel()

# convert it to a matrix
vector_circle = np.matrix(flat_arr_circle)

# do something to the vector
vector_circle[:,::10] = 128

# reform a numpy array of the original shape
arr2_circle = np.asarray(vector_circle).reshape(shape_circle)

# make a PIL image
img2_circle = Image.fromarray(arr2_circle, 'RGBA')
img2_circle.show()
```

```
from Pyfhel import Pyfhel, PyCtxt

he = Pyfhel()        # Object in charge of all homomorphic operations
he.contextGen(255) # Choose the maximum size of your integers
he.keyGen()     # Generate your private/public key pair


```

```
# Now to the encryption/operation/decryption

encrypted_circle = []
for n in flat_arr_circle:
    encrypted_circle.append(he.encryptInt(n))
```

```


encrypted_square = []
for i in flat_arr_square:
    encrypted_square.append(he.encryptInt(i))
```

```
len(encrypted_square)
```

```
len(encrypted_circle)
```

```
addingencrypted=[]
for i in range(len(encrypted_square)):
  addingencrypted.append(he.add(encrypted_square[i],encrypted_circle[i]))
  
```

```
addeddecrypted=[]
for i in range(len(addingencrypted)):
  addeddecrypted.append(he.decryptInt(addingencrypted[i]))
  
```

```
addeddecrypted=np.array(addeddecrypted, dtype='uint8')
arr2 = np.asarray(addeddecrypted).reshape(shape_square)

# make a PIL image
img2 = Image.fromarray(arr2, 'RGBA')
img2.show()
```
